## 1. Finding the Pooled Variance

To begin finding the $t$ statistic first we need to find the pooled variance between the two conditions which will be used in the estimated standard error. Pooled variance is the average of both group's variances $S^2$ with the larger groups variance carrying the most weight. Pooled variance is used when both group's sample size $n$ are not the same, later we will discuss what to do when the sample sizes are the same. The sum of squares of group 1 is added to to the sum of squares for group 2 divided by the degrees of freedom of group 1 plus the degrees of freedom of group 2.

### Formulas within Pooled Variance
$SS = \Sigma(X - M)^2$<br>$df = n - 1$

### Pooled Variance Formula
$S^2_{P} = \frac{SS_{1} + SS_{2}}{df_{1} + df_{2}}$

## 2. Calculating the Estimated Standard Error

The estimated standard error shows the difference between the sample means and its representation of the difference between two population means. In this formula, the pooled variance we calulated is used in the numerator of the fraction.

### a. When Sample Sizes are Unequal

$S_{M_{1}-M_{2}} = \sqrt{\frac{S^2_{P}}{n_{1}} + \frac{S^2_{P}}{n_{2}}}$

### b. When Sample Sizes are Equal
$S_{M_{1}-M_{2}} = \sqrt{\frac{S^2_{1}}{n_{1}} + \frac{S^2_{2}}{n_{2}}}$

As mentioned before, there is a different formula used to calculate estimated standard error $S_{M_{1}-M_{2}}$ for samples with equal sizes. In this formula, the calculated pooled variance is replaced by the variance $S^2$ of each sample.

## 3. Calculate t-Statistic

Finally we have all the information needed to calculate our t-statistic. The formula requires the sample means for each sample, the population means of each population (usualy 0 unless the population is found or given), and the estimated standard error.

### t-Statistic Formula
$t = \frac{(M_{1}-M_{2}) - (\mu_{1}-\mu_{2})}{S_{M_{1}-M_{2}}}$

## 4. Make a decision

After the t-statistic has been calculated, we compare it to the critical value. We find the critical value by setting an $\alpha$ level or the significance level, which is the probability of rejecting the null hypothesis when it is true. The $\alpha$ is commonly set to $.05$ or $.1$. We then look at a t table which lists all the critical values for different degrees of freedoms, different $\alpha$ levels, and the test at one-tail or two-tailed.

### t Table


```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 2*sigma, mu + 2*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

This is an example of a normal distribution which can be used to better explain the concept of critical values and critical regions. For example, if the critical t found from the t-table is $\pm1.5$, all $t$ values that fall between $-1.5$ to $+1.5$ are within the critical region. Therefore we would accept the null hypothesis $H_{0}$ which states that there is no difference between the two samples. But, if the $t$ statistic falls outside of the critical region then we reject the null hypothesis and accept the alternative hypothesis $H_{1}$ which states that there is a difference between the samples. From this we can draw a conclusion to the significance of the results.
